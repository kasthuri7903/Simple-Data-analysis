## Merge all the sales data into a single csv file

```python
files = [file for file in os.listdir(r'C:\Users\kasth\OneDrive\Desktop\keith_sales_Data')]
```

```python
all_months_data = pd.DataFrame()
all_months_data
```

```python
for file in files:
    df = pd.read_csv(r'C:\Users\kasth\OneDrive\Desktop\keith_sales_Data'+"/"+file)
    all_months_data = pd.concat([all_months_data,df])


all_months_data.to_csv("all_data.csv",index=False)
```
